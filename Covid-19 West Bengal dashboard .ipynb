```python
import dash
import dash_html_components as html  
import dash_core_components as dcc
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
from dash.dependencies import Input, Output

app = dash.Dash()


covid=pd.read_excel(r'C:\Users\Sanjoy Das\Downloads\district_wise_cases_of_west_bengal.xlsx')


fig1 = go.Figure()
fig2 = go.Figure()
fig3 = go.Figure()
fig4 = go.Figure()


app.layout=html.Div([
                   
                   html.Div(dcc.Dropdown(id='dropdown1',options=[{'label':'Barplot','value':'bp'},
                                                                 {'label':'Scatterplot','value':'sp'}],
                                                                placeholder="Select an Option",
                                                                style={'width':'49%'})),
                   html.Div([dcc.Graph(id='figure1',figure=fig1,style={'width':'49%','display':'inline-block'}),
                             dcc.Graph(id='figure2',figure=fig2,style={'width':'49%','display':'inline-block'})]),
    
                   html.Div([dcc.Graph(id='figure3',figure=fig3,style={'width':'49%','display':'inline-block'}),
                             dcc.Graph(id='figure4',figure=fig4,style={'width':'49%','display':'inline-block'})])
                    ]) 


@app.callback([Output('figure1','figure'),Output('figure2','figure'),Output('figure3','figure'),Output('figure4','figure')],[Input('dropdown1','value')])

def updatefigure(dropdown1):
    global fig1,fig2
    if (dropdown1=='bp'):
        
    fig1=go.Figure(px.bar(covid,x="District",y="Active _Cases",animation_frame="Month",title="Covid-19 Active Cases over 28th May to 28th October"))
fig1.update_layout(title="Animated Bar Plot-Covid-19 Active Cases in West Bengal over 28th May to 28th October",xaxis_title="District",yaxis_title="Active Cases")
               
        fig2=go.Figure()

    if (dropdown1=='sp'):

        fig1=go.Figure()
               
    fig2=go.Figure(px.scatter(covid,y="District",x="Active _Cases",animation_frame="Month",size='Active _Cases',color='Active _Cases',hover_name='District',title='Animated Scatter Plot of Active cases in West Bengal[28th May-28th October]'))
fig2.update_layout(title="Animated Scatter Plot of Active Cases in West Bengal[28th May-28th October]",xaxis_title="Active Cases",yaxis_title="District")
    return fig1,fig2
        


if __name__ == '__main__':         
    app.run_server()
```
